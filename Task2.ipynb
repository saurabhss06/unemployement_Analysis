```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
unempinfo = pd.read_csv("Unemployment.csv")
unempinfo
```

**Visualization**

```python
unempinfo=pd.DataFrame(unempinfo)
y=unempinfo[' Estimated Unemployment Rate (%)']
x=unempinfo['Region']
pr= pt.figure(figsize=(40, 15))
pt.xlabel("Indian States",fontweight='bold',fontsize=20)
pt.ylabel("Unemployment rate in %",fontweight='bold',fontsize=20)
pt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
pt.bar(x,y, color='#34e8eb',align='center')
```

**State wise rate of unemplyement**

```python
u_emp= unempinfo[['Region',' Estimated Unemployment Rate (%)']].groupby('Region').sum().sort_values(by=' Estimated Unemployment Rate (%)', ascending =False)
u_emp
```

```python
pt.figure(figsize=(12,10))
pt.title('State Wise Unemployment In India ')
sb.histplot(x=' Estimated Unemployment Rate (%)', hue="Region", data=unempinfo)
pt.show()
```

**Unemployement in INDIA**

```python
pt.figure(figsize=(12,10))
pt.title('Unemployment In India')
sb.histplot(x=' Estimated Unemployment Rate (%)', hue="Region.1", data=unempinfo)
pt.show()
```
